```
#Lecture fichier
dating = pd.read_csv('Speed_dating.csv', encoding="ISO-8859-1")
dating.head(10)
```

```
dating.columns
```

```
# Selection des colonnes à conserver
# colonne a garder :
dating = dating.loc[:, ["iid", "pid", "partner","gender", "age", "age_o", "race", "race_o", "attr_o", "samerace", 
                         "imprace", "imprelig", "match", "int_corr", "field", "field_cd", "from", "zipcode", "income", 
                         "goal", "date", "go_out", "career", "career_c", "tvsports", "exercise", "dining", "museums", 
                         "art", "hiking", "gaming", "clubbing", "reading", "tv", "theater", "movies", "concerts", "music",
                         "shopping", "yoga", "exphappy", "expnum", "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", 
                         "shar1_1", "attr2_1", "sinc2_1", "intel2_1", "fun2_1", "amb2_1", "shar2_1", "attr3_1", "sinc3_1"]]
dating.head(10)
```

```
dating.isnull().sum()
```

```
pd.DataFrame(dating.isnull().sum(), columns= ["Valeur NAN"])
```

```
#Suppression de colonnes dont la valeur des NAN est bcp trop élever aux risque de fauser nos résultats
dating = dating.drop(["expnum","income"], axis=1)
dating.isnull().sum()
```

```
corr = dating.corr()
plt.figure(figsize=(30,30))
sns.heatmap(corr, annot=True)
```

```
#Nombre de participant
#Nombre de Rdv
nbre_participant = len(dating['iid'].unique())
print("Le nombre de personnes participant aux rdv est de {} ".format(nbre_participant))

nbre_rdv = len(dating.index)
print("Le nombre de rdv est de {} ".format(nbre_rdv))
```

```
# La répartition selon l'age
age = dating[np.isfinite(dating['age'])]['age']
plt.hist(age.values, color="red")
plt.xlabel('Age')
plt.ylabel('participant')
```
